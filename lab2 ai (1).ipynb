ASTHAMA DISEASE DATASET

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filepath=r"C:\Users\ASUS\Downloads\asthma_disease_data.csv"
df=pd.read_csv(filepath,encoding='latin-1')
```

```python
df.head()
```

COUNTPLOT

```python
gender_mapping={0:'Female',1:'Male'}
df['gender_label']=df['Gender'].map(gender_mapping)
sns.countplot(data=df,x='Gender')
gender_counts = df['gender_label'].value_counts()
gender_counts.plot(kind='bar',rot=0,width=0.5)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()


```

Histogram

```python
plt.title("AGE DISTRIBUTION")
plt.hist(df['Age'],color="lightblue")
```

```python

plt.title("Gender Distribution")

plt.pie(df['Gender'].value_counts(),autopct='%0.2f')

```

```python
plt.title('Boxplot of Education Label')
sns.boxplot(df['EducationLevel'])
df.describe()
```

```python
sns.lineplot(x='EducationLevel',y='Gender',data=df,c='#0f0',lw=1)
plt.title('LINEPLOT')
```

```python
sns.kdeplot(df['LungFunctionFVC'])
```

PairPlots

```python
sns.barplot(x="Gender",y="EducationLevel",data=df,palette="Set2")
plt.title('Education Level by Gender')
```

```python
sns.boxplot(x='Smoking', y='BMI', data=df, hue='Gender')
plt.xlabel('Smoking Status')
plt.ylabel('BMI')
plt.title('Boxplot of BMI by Smoking Status and Gender')
plt.show()
```

```python
sns.scatterplot(x='Age', y='BMI', data=df)
plt.xlabel('Age')
plt.ylabel('BMI')
plt.title('Scatter Plot of Age vs. BMI')
plt.show()
```

```python
sns.displot(df['LungFunctionFVC'])
```
