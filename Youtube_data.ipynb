```python
category_suscriibers_mean=df.groupby("category")["subscribers"].mean().sort_values(ascending=False)
category_suscriibers_mean
```

```python
category_suscriibers_mean.plot(kind="bar")
plt.title("Mean Suscribers by Category")
plt.xlabel("Category")
plt.ylabel("Mean Suscribers")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.boxplot(x="channel_type",y="highest_monthly_earnings",data=df)
plt.title("Distribution of Highest Monthly Earning by Channel Type")
plt.xticks(rotation=90)
plt.show()
```

```python
df.groupby(["category","Country"])["subscribers"].sum().unstack().plot(kind="bar",stacked=True)
plt.title("Suscribers Breakdown by Category and Country")
plt.xlabel("Category")
plt.ylabel("Total Suscribers")
plt.legend(title="Country",bbox_to_anchor=(1.05,1),loc="upper left")
```

```python
plt.scatter(df["Gross tertiary education enrollment (%)"],df["subscribers"],alpha=0.5)
plt.title("scatter plot of subscribers vs gross tertiary education enrollment")
plt.xlabel("Gross tertiary education enrollment")
plt.ylabel("Subscribers")
plt.show()
```

```python
df.groupby("channel_type")["Unemployment rate"].mean().plot(kind="bar")
plt.title("Average Unemployment Rate by Channel Type")
plt.xlabel("Channel Type")
plt.ylabel("Average Unemployment Rate")
plt.xticks(rotation=90)
plt.show()
```

```python
df.groupby(["created_year","channel_type"])["subscribers"].mean().unstack().tail().plot(kind="bar",width=1)
plt.title("Average Suscribers by Create Tear and Channel Type ")
plt.xlabel("Created Year")
plt.ylabel("Average_Suscribers")
plt.xticks(rotation=45)
plt.legend(title="Channel Type",bbox_to_anchor=(1.05,1),loc="upper left")
plt.show()
```

```python
df.groupby(["created_year","created_month"])["subscribers"].sum().unstack().head().plot(kind="area",stacked=True)
plt.title("Subscribers Growth Over Time")
plt.xlabel("Year")
plt.ylabel("Total Subscribers")
plt.legend(title="Month",bbox_to_anchor=(1.05,1),loc="upper left")
plt.xticks(rotation=45)
plt.show()
```

```python
df.groupby(["created_year","Country"])["highest_yearly_earnings"].mean().unstack().head().plot(kind="line",marker="o")
plt.title("Trend of Highest Yearly Earnings by Created Year and Country")
plt.xlabel("Created year")
plt.ylabel("Average Highest Yearly Earnings")
plt.legend(title="Country",bbox_to_anchor=(1.05,1),loc="upper left")
plt.xticks(rotation=45)
plt.show()
```

```python
channel_created_suscribers_mean=df.groupby(["channel_type","created_year"])["subscribers_for_last_30_days"].mean().unstack()
channel_created_suscribers_mean
```

```python
sns.heatmap(channel_created_suscribers_mean,cmap="YlGnBu",annot=True)
plt.title("Mean subscribe for last 30 days by channel Type and created Year")
plt.xlabel("Created year")
plt.ylabel("Channel type")
plt.show()
```

```python
df.groupby("created_month")["subscribers_for_last_30_days"].mean().plot(kind="line",marker="o")
plt.title("Average Subscribers for last 30 days by month")
plt.xlabel("Month")
plt.ylabel("Average Subscribers for last 30 days")
plt.show()
```
